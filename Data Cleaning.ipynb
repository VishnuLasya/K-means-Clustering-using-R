```python
distances=[]

K=range(1,10)
for k in K:
    ClusterInfo = kmeanModel =KMeans(n_clusters=k).fit(df)
    distances.append(ClusterInfo.inertia_)
    
plt.plot(K,distances,'bo-')
plt.xlabel('K-clusters')
plt.ylabel('Distance')
plt.title('Cluster Values and Distances')
plt.show()
```
